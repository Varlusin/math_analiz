## Դիֆերենցիալ հաշվի հիմնական թյորեմները։

### Ֆերմաի թեորեմը։

 Եթե $f$ ֆունկցիան $X$ միջակայքի $x_0$ ներքին կետում ընդունում է իր մեծագույն (փոքրագույն) արժեքը և գոյություն ունի $f'(x_0)$-ն, ապա $f'(x_0) = 0$:

եթե հիշենք ածանցիալի երկրաչափական իմաստը Ֆերմաի թեորեմը ասում է որ ֆունկցիաի գրաֆիկին տարած շոշափողը ֆունկցիայի  max  (min) կետերում զուգահեռ է $x$-երի առանցքին։  

### Ռոլի թեորեմը։

Դիցուկ $[a,b]$ միջակայքում որոշված ֆունկցիան բավարարում է հետևյալ պայմաններին՝

1. Աննդհատ է $[a,b]$ ում։
1. ունի ածանցիալ առնվազն $(a,b)$ միջակայքում։
1. $f(a) = f(b)$

$[a,b]$ միջակայքում գոյություն ունի $c$ կետ այնպիսին որ $f'(c) = 0$

### Լագրանժի վերջաոր աճերի բանաձևը։

Եթե $f(x)$ ֆունկցիան որոշված և աննդհատ է $[a,b]$ հատվածում, առնվազն դիֆերենցելի է $(a,b)$ ում ապա գոյություն ունի $c ∈ [a,b]$   ներքին կետ այնպիսին որ՝
$$\frac {f(b)- f(a)}{b-a} = f'(c)$$

Թեորեմը երկրաչափորեն նշանակում է որ $[a,b]$ հատվածում աննդհատ ֆունկցիաի ծայրակետերը իրար միացնող ուղղին զուգահեռ ուղիղ կլինի այնպիսին որ, անպայման կշոշափի ֆունկցիաի գրաֆիկին $[a,b]$ հատվածում։

Որպես $f(x)$ ֆունկցիա դիտարկենք $f(x) = 3 + 3 x \sin{(x)}$ Աննդհատ ֆունկցիան $[1,6]$ միջակայքում։  

```python
import numpy as np
import matplotlib.pyplot as plt
x = np.linspace(1, 6 , 100)
y= 3 + 3*x*np.sin(x)
plt.plot(x , y)
plt.scatter([1, 6],[y[0], y[-1]])
plt.plot([1, 6],[y[0], y[-1]])
plt.text(4.5, 7.5, r'$\frac{f(b)- f(a)}{b-a} = -1.51 ...$')
plt.xlabel("X")
plt.ylabel("$f(x) = 3 + 3 x \sin{(x)}$")
plt.title("ՖՈՒՆԿՑԻԱՅԻ ԳՐԱՖԻԿԸ")
plt.grid(True)
```

Ահա այսպիսին է ֆունկցիաի տեսքը նշված միջակայքում։ Այժմ անհրաժեշտ է գտնել այն $c ∈ [a,b]$ թվերից գոնե մեկը, որ $f'(c) = \frac{f(b)- f(a)}{b-a}$: ֆունկցիաի ածանցիալը հաշվելու համար օգտվենք շատ պարզ ալգորիթմից, որի հիմքում ընկած է ածանցիալի սահմանումը։ Ալգորիթմը թույլ է տալիս գնահատել ածիանցիալ ֆունկցիայի արժեքը՝ այսպիսով ալգորիթմը սկսում է $a$ ծայրակետից այդ կետում $x$- ին տալիս է որոշակի աճ (որքան փոքր լինի այդ աճը այնքան ավելի ճշգրիտ կլինի գնահատականը) հաշվում է ֆունկցիաի աճի և $x$ - ի աճի հարաբերությունը, որը համեմատում է անհրաժեշտ արժեքի հետ, եթե համեմատության արդյունքը բավարարում է ցիկլը կանգ է առնում հակառակ դեպքում շարունակում է նույն գործողությունը $x$- նոր արժեքով։  

```python
c = 1
dx = 0.00001
delta_x = 0.0001
find_deriv = (y[-1] - y[0])/5
while True:
  deriv = 3*((c+dx)*np.sin(c+dx)-c*np.sin(c))/dx
  if abs(find_deriv - deriv) <= 0.001:
    break
  else:
    c = c + delta_x

b = 3 + 3*c*np.sin(c) - find_deriv*c
x = np.linspace(1, 6 , 100)
y= 3 + 3*x*np.sin(x)

#plt.figure(figsize = (10, 5))
plt.plot(x , y, color = 'b')
plt.scatter([1 , 6 , c],[y[0] , y[-1] , 3 + 3*c*np.sin(c)] , color = 'g')
plt.plot([1, 6],[y[0], y[-1]], color = 'b')
plt.plot([1, 6], [b + find_deriv , b + 6*find_deriv ], color = 'r')
plt.xlabel("X")
plt.ylabel("$f(x) = 3 + 3 x \sin{(x)}$")
plt.title("ՖՈՒՆԿՑԻԱՅԻ ԳՐԱՖԻԿԸ")
plt.grid(True)
```

Թեորեմը ունի ապացույց սակայն այն փաստը որ ցիկլը կանգ է առնում դա էլ կարող է հանդիսանալ որպես այլընտրանք ապացույց։

### Կոշիի վերջաոր աճերի բանաձևը։

Եթե $f$ և $g$ ֆունկցիաները աննդհատ են $[a,b]$ հատվածում և դիֆերորենցելի են $(a,b)$ միջակայքում, ընդ որում՝ $x \in (a,b)$ միջակայքում $g'(x) \neq 0$, ապա գոյություն ունի $c \in (a,b)$ այնպիսին, որ
$$ \frac {f(b) - f(a)}{g(b) - g(a)} = \frac {f'(c)}{g'(c)}$$

### Դարբուի թեորեմը։

Եթե $f(x)$ ֆունկցիան դիֆերենցելի է $[a,b]$ հատվածում, և $f'(a)f'(b)< 0$ ապա գոյություն ունի $c \in (a,b)$ այնպիսին որ $f'(c) = 0$

__Լրացում__

Եթե $f(x)$ ֆունկցիան դիֆերենցելի է $[a,b]$ հատվածում և $f'(a) < C < f'(b)$ ապա գոյություն ունի $c \in (a,b)$ այնպիսին որ $f'(c) = C$
