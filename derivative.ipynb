## ՖՈԻՆԿՑԻԱԻ ԱԾԱՆՑԻԱԼ։

```python
import numpy as np
import matplotlib.pyplot as plt
```

Դիցուկ ունենք որևէ $f(x)$ մեկ փոփոխականի ֆունկցիա։ Տրված $M(X=x,Y=f(x))$ կետում  տարված շոշափող դեռևս դպրոցական տարիներից հայտնի սահմանումով ասվում է այն ուղիղն է որը այդ կետը հատում է մեկ անդամ։ Իսկապես այս սահմանումը լիարժեք չէ։ Որպիսի ավելի լավ հասկանանք որպես $f(x)$ դիտարկենք $x^2$ պարզագույն ֆունքցիան։

```python
x=np.arange(-10,10, 0.1)
y=x**2
x_tangential = np.array([2.5,7.5])
y_tangential = 10*x_tangential-25
yNot_tan = 30*x_tangential-125
plt.plot(x,y)
plt.plot(x_tangential, y_tangential)
plt.plot(x_tangential, yNot_tan)
plt.annotate('$M(5,25)$', xy=(5.1, 24.9), xytext=(7.5, 25),
            arrowprops=dict(facecolor='black', shrink=0.1))
plt.xlabel("X")
plt.ylabel("$f(X)={x^2}$")
plt.title("ՖՈՒՆԿՑԻԱՅԻ ԳՐԱՖԻԿԸ։")
plt.axhline(y=0, color='k')
plt.axvline(x=0, color='k')
plt.scatter(x=5 , y=5**2)
plt.grid(True)
```

Ինչպես տեսնում եմք մեր իմացած սահմանումը թերի է և կարիք ունի ճշտգրման քանի որ ցանկացած $M(X,Y)$ կետում կարելի է տանել անվերջ թվով ուղիղներ որոնք կհատվեն ընդամենը մեկ անգամ։ Ուստի անհրաժեշտ է տալ ավելի ընդհանուր սահմանում։

Ընդհանուր առմամբ տալ ուղիղ գծի հավասարում նշանակում է իմանալ ${y = a + bx}$ հավասարման  a և b գործակիցները։ Դժվար չի լինի իմանալ a֊ն եթե ունենանք b-ն քանի եր ուղիղ գիծը պետք է անցնի M կետով այդտեղից էլ կարող ենք գտնել a գործակիցը։

Այսպիսով կորի վրա դիտարկենք $M_0(X_0,Y_0)$ և $M_1(X_1,Y_1)$ կետերը և այդ երկու կետերը իրար միացնող ուղիղը։

```python
x=np.arange( 0,10, 0.01)
y=x**2
plt.plot(x,y)

plt.text(4.2,15, "${M_0(4,16)}$")
plt.text(9.2, 80, "${M_1(9,81)}$")
plt.xlabel("X")
plt.ylabel("$f(X)={x^2}$")
plt.title("ՖՈՒՆԿՑԻԱՅԻ ԳՐԱՖԻԿԸ։")
plt.axhline(y=0, color='k')
plt.axvline(x=0, color='k')
plt.scatter([4,9],[16,81])
plt.plot([4,9],[16,81])
plt.grid(True)
```

Ստացված կորի անկյունային գործակիցը որոշվում է ${M_0 , M_1}$ կետերի համապատասխան կորդինատների տարբերության հարաբերությամբ։

նշանակենք՝ ${Δx=X_1-X_0}$ և ${Δy=Y_1-Y_0}$

${b= \frac{Δy}{Δx}}$

Այժմ երկու կետերը մոտեցնենք իրար։

```python
x=np.arange(3.5,9.5, 0.01)
y=x**2
plt.plot(x,y)
x_m= np.array([4,5,6,7,8,9])
y_m = x_m**2
plt.xlabel("X")
plt.ylabel("$f(X)={x^2}$")
plt.title("ՖՈՒՆԿՑԻԱՅԻ ԳՐԱՖԻԿԸ։")
plt.scatter(x_m,y_m)
for i in range(1,6,1):
  plt.plot([4,x_m[i]], [16, y_m[i]])
plt.grid(True)
```

Ահա երբ մոտենում են երկու կետերը մենք մոտենում ենք շոշափողին։ այժմ հաշվենք հարաբերության սահմանը երբ $Δx$֊ը ձգտում է 0֊ի։

$\displaystyle \lim_{Δx \to \ 0} \frac{Δy}{Δx} = \lim_{Δx \to \ 0} \frac{(X_0+ΔX)^2-X_0^2}{Δx} = \lim_{Δx \to \ 0} \frac{2ΔxX_0+Δx^2}{Δx} = 2X_0$

Որպես շոշափող կհամարենք հենց ${M_0 M_1}$  անցմող ուղիղը երբ ${Δx}$ կձկտի 0-ի: Իսկ անկյունային գործակիցը «X առանցքի եվ շոշափողի կազմած անկյան տանգ.» կլինի մեր հաշված սահմանը:

## ածանցիալի սահմամումը:

Դիցուկ ${y=f(x)}$ ֆունկցիան որոշված է ${X}$ միջակայքում այդ միջակայքում վերցնենք որեվէ ${x_0}$ կեկ այդ կետում ${x}$ ին տանք այնպիսի ${Δx}$ աճ որ ${x_0 + Δx}$-ը նույնպես պատկանի ${X}$ միջակայքին: Ֆունկցիան կունենա ${Δy=f(x_0+Δx)-f(x_0)}$ աճ: ${f(x)}$ ֆունկցիաի ածանցիալ ${x_0}$ կետում կանվանենք՝

$$\frac{dx}{dy}= \lim_{Δx \to \ 0} \frac{f(x_0 + Δx) - f(x_0)}{Δx}$$






### ՊԱՐԶ ՖՈԼՆԿՑԻԱՆԵՐԻ ԱԾԱՆՑԻԱԼՆԵՐ։

1. ${ f(x)= C}$ :  $\frac{df(x)}{dx}= 0:$
1. ${f(x)= x^a}$:  $\frac{df(x)}{dx}= a x^{a-1}:$
1. ${f(x) = a^x}$: $\frac{df(x)}{dx}= a^{x}\ln{a}:$
1. $f(x) = \log_a x$: $\frac{df(x)}{dx}= \frac{1}{x}{log_a e}:$
1. $f(x) = sin(x)$ : $\frac{df(x)}{dx} = cos(x)$
1. $f(x) = cos(x)$ : $\frac{df(x)}{dx} = -sin(x)$
1. $f(x) = tg(x)$ : $\frac{df(x)}{dx} = \frac{1}{cos^2(x)}$
1. $f(x) = ctg(x)$ : $\frac{df(x)}{dx} = -\frac{1}{sin^2(x)}$




### ՖՈՒՆԿՑԻԱԻ ԱՃԻ ԲԱՆԱՁևԸ։
Դիցուկ $f(x)$ ֆունկցիան որոշված է ${X}$ միջակայքում ${x_0}$ կետը պատկանում է ${X}$ միջակայքին։ ${x_0}$ կետում ունի վերջաոր ածանցիալ։

$${f'(x_0)}= \frac{f(x_0 + Δx) - f(x_0)}{Δx}, \text{ երբ }  {Δx\to 0}  $$

Նշանակենք՝

$$a = \frac{f(x) - f(x_0)}{x-x_0}-f'(x_0)$$
Կունենանք՝

$${f(x_0 + Δx) - f(x_0)} = Δx f'(x_0) + Δx a$$

${a}$-ն ֆունկցիա է ${Δx}$ ից և երբ ${Δx\to 0}$ ի ${a\to 0}$

 __հետևանք՝__
Եթե $f$ ֆունկցիան ունի $x_o$ կետում վերջաոր ածանցիալ՝ ապա այդ կետում $f$ ֆունկցիան աննդհատ է։ Հակառակ պնդումը ճիշտ չէ։

Օրինակ՝ $f(x) = |x|$ ֆունկցիան $x = 0 $ կետում երկու կողմից ունի ներջաոր ածանցիալ, բայց աննդհատ չէ։ Դիտարկենք $f(x) = x sin(\frac{1}{x})$ ֆունկցիան։

```python
x = np.linspace(-np.pi, np.pi, 500)
y = x* np.sin((1/x))
plt.xlabel("X")
plt.ylabel("$f(X)$")
plt.title(" ՖՈՒՆԿՑԻԱՅԻ ԳՐԱՖԻԿԸ:")
plt.plot(x, y)
plt.grid(True)
```

Այս ֆունկցիան 0֊ու աննդհատ է բայց ածանցիալ չունի։

### ԱԾԱՆՑՄԱՆ ԿԱՆՈՆՆԵՐԸ։

դիցուկ $u$ և $v$ ֆունկցիաները որոշված են և ունեն վերջաոր ածանցիալ $x_0$ կետում ապա՝
1. $(Cu)' = Cu'$
1. $(u ± v)' =  u' \pm v'$
1. $(uv)' = u' + v' $
1. $(\frac{u}{v})' = \frac{u'v - v'u}{v^2}$

### Բարդ ֆունկցիաի ածանցիալը։

Դիցուկ տրված է $F(x) = f(\varphi (x))$  ֆունկցիան $φ (x)$ ֊ ը $x_0$ կետում որոշված է $t_0 = \varphi(x_0)$ և ունի վերջաոր ածանցիալ, իսկ  $f(t_0)$ կետում որոշված է և ունի վերջաոր ածանցիալ՝

$F'(x_0) = f'(\varphi (x_0)) \varphi ' (x_0) $

### Հակադարձ ֆունկցիաի ածանցիալը՝

Դիցուկ $y=f(x)$֊ ֆունկցիան որոշված, խիստ մոնոտոն և աննդհատ է $X$ միջակայքում՝ ըստ հակադարձ ֆունկցիայի գոյություն ունենալու թյորեմի $x=g(y)$ հակադարձ ֆունկցիան որոշված է $Y$ միջակայքում։

### Թյորեմ՝
եթե $f(x)$- ֆունկցիան $x_0$ կետում ունի վերջաոր ածանցիալ $f'(x_0) \neq 0$ `

$g'(x_0) = \frac{1}{f'(x_0)}$

Այն ունի երկրաչափական պարզ իմաստ, եթե հիշոնք որ ֆունկցիաի ածանցիալը կորին տարած շոշափողի և $x$֊ի առանցքի հետ կազմած անկյան $tn$֊է նշանակում է հակադարձ ֆունկցիաի ածանցիալը շոշափողի և $y$֊ի հետ կազմած անկյան $tn$֊է։

__Օրինակ՝__
1. $x= sin(y) \text{,   } y= arcsin(x) \text{,   } -\frac{\pi}{2} < y < \frac{\pi}{2} \text{,   } y'_x= \frac{1}{x'_y} = \frac{1}{cos(x)} = \frac{1}{\sqrt{1-x^2}} $
1. $x= cos(y) \text{,   } y= arccos(x) \text{,   } 0 < y < \pi \text{,   } y'_x= \frac{1}{x'_y} = -\frac{1}{sin(x)} = -\frac{1}{\sqrt{1-x^2}} $

1. $x= tn(y) \text{,   } y= arctn(x) \text{,   } -\frac{\pi}{2} < y < \frac{\pi}{2} \text{,   } y'_x= \frac{1}{x'_y} = cos^2 y = \frac{1}{1+x^2} $

1. $x= ctn(y) \text{,   } y= arcctn(x) \text{,   } 0 < y < \pi \text{,   } y'_x= \frac{1}{x'_y} = - sin^2 y = -\frac{1}{1+x^2} $
