## ԴԻՖԵՐԵՆՑԻԱԼ

Դիցուկ տրված է $f(x)$ ֆունկցիան որը որոշված է $X$ միջակայքում։ և աննդհատ է $x_0$ կետում: Այդ կետում տանք $Δx$ աճ։ Ֆունկցիան կունենա $Δy$ աճ։ Կասենք որ ֆունկցիան դիֆերենցելի է $x_0$ կետում եթե՝
$$f(x_0 + Δx) - f(x_0) = A Δx + α Δx \text{ (1)}$$

որտեղ $A$֊ն կախված չի $Δx$֊ից իսկ $α$-ն $Δx$ ից կաժխված ֆունկցիա է և անվերջ փոքր է երբ $Δ x \to 0$:

Այժմ ցուից տանք որ՝ $A = f'(x_0)$։

$$ \frac{Δf}{\Delta x} = A + α ,\text{ երբ } Δx \to 0 ,  α \to 0 $$

երկու կողմում անցնելով սահմանի կունենանք՝ $A = f'(x_0)$: $(1)$ արտահայտության մեջ $A Δx$ որը գծաին է կախված  $Δ x$֊ից կոչվում է գլխաոր մաս։

__ՍԱՀՄԱՆՈՒՄ__

$x_0$ կետում դիֆերենցելի $ f(x)$ ֆունկցիաի աճի գլխաոր մասը կոչվում է ֆունկցիաի դիֆերենցիալ այդ կետում և նշանակվում է $df(x_0)$  կամ $dy(x_0)$ սիմվոլով։

Ֆունկցիաի դիֆերենցիալն ունի պարզ երկրաչափական իմաստ։



```python
import numpy as np
import matplotlib.pyplot as plt
x = np.linspace(1, 5 , 100)
y= x**2
plt.plot(x , y)
plt.plot([1,4],[0,12])
plt.plot([2,4,4], [4,4,16], "--", color = 'k')
plt.scatter([2,4, 4, 4],[4,16, 4, 12])
plt.text(2.9,3, "$\Delta x = 2$")
plt.text(4.1,7.5, "$df(x_0) \Delta x$")
plt.text(4.1,14, "$α \Delta x$")
plt.grid(True)
```

Նկարում պատկերված է $f(x) = x^2$ ֆունկցիան։ Ֆունկցիաի գրաֆիկին  $x_0 = 2$ կետում  տանենք շոշափող։ Երբ $x$ ֊ին $x_0$ կետում տանք $Δx$ աճ ֆունկցիան կունենա $Δf$ աճ՝ որը կարելի է ներկայացնել երկու գումարելիներով որից մեկը շոշափողի ֆունկցիաի աճն է, որն էլ ֆունկցիաի դիֆերենցիալն է $x_0$ կետում։

__ԴԻՖԵՐԵՆՑՄԱՆ ԿԱՆՈՆՆԵՐԸ։__

Դիցուկ $u$ և $v$ ֆունկցիաները դիֆերենցելի է $x_0$ կետում ապա՝
1. ${\partial (cu)} = {c∂u}$

1. ${\partial (u {±} v)} = {∂u \pm ∂ v}$

1. ${\partial (uv)} = {v∂u + u ∂v}$

1. ${\partial (\frac {u}{v})} = {\frac {v \partial u - u \partial v}{v^2}}$
